```python
from ultralytics import YOLO

# Load the checkpoint
model = YOLO('/content/drive/MyDrive/yolov8_training_checkpoints/yolov8n_training/weights/last.pt')

# Resume training
results = model.train(
    data="/content/drive/MyDrive/yolov8_dataset/data.yaml",
    epochs=100,  # Continue till 100
    batch=16,
    imgsz=640,
    device="cuda",
    save=True,
    project="/content/drive/MyDrive/yolov8_training_checkpoints",
    name="yolov8n_training",
    resume=True  # Resume from last.pt
)
```
