```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the path to your image in Google Drive
image_path = "/content/drive/MyDrive/images/1000check.jpg"  # Update with your actual path

def preprocess_image(image_path):
    # Load image directly in grayscale
    gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if gray is None:
        raise ValueError("Error: Image not found or invalid path!")

    # Resize to 640x640
    resized = cv2.resize(gray, (640, 640))

    # Apply Gaussian Blur
    blurred = cv2.GaussianBlur(resized, (5, 5), 0)

    # Perform Canny Edge Detection
    edges = cv2.Canny(blurred, 50, 150)

    return resized, blurred, edges

def show_images(images, titles):
    """ Displays images using Matplotlib """
    plt.figure(figsize=(12, 4))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img, cmap="gray")
        plt.title(title)
        plt.axis("off")
    plt.show()

# Preprocess image
resized, blurred, edges = preprocess_image(image_path)

# Display images
show_images([resized, blurred, edges], ["Resized (Grayscale)", "Blurred", "Canny Edges"])

# Save processed images back to Google Drive
output_dir = "/content/drive/MyDrive/Processed_Images/"
cv2.imwrite(output_dir + "resized.jpg", resized)
cv2.imwrite(output_dir + "blurred.jpg", blurred)
cv2.imwrite(output_dir + "edges.jpg", edges)

print(f"Processed images saved in: {output_dir}")
```

```python
import os
import cv2
import time

# Start timer
start_time = time.time()

# Initialize a counter for processed images
processed_count = 0

# Define the folder path for your images
image_folder_path = "/content/drive/MyDrive/yolov8_dataset/test/images"  # Update with the actual path to your image folder
output_dir = "/content/drive/MyDrive/Preprocessed_Images/"  # Update with the output path for processed images

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

def preprocess_image(image_path):
    """
    Preprocesses an image step by step: Resize -> Grayscale -> Gaussian Blur -> Canny Edge Detection.

    Args:
        image_path (str): Path to the input image.

    Returns:
        tuple: (resized, grayscale, blurred, edges)
    """
    # Load image directly in grayscale mode
    grayscale = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if grayscale is None:
        raise ValueError(f"Error: Image not found or invalid path for {image_path}!")

    # Resize to 640x640
    resized = cv2.resize(grayscale, (640, 640))

    # Apply Gaussian Blur
    blurred = cv2.GaussianBlur(resized, (5, 5), 0)

    # Perform Canny Edge Detection
    edges = cv2.Canny(blurred, 50, 150)

    # Debugging: Print the dimensions of the processed images
    print(f"Returning 3 processed images for {image_path}: resized {resized.shape}, blurred {blurred.shape}, edges {edges.shape}")
    return resized, blurred, edges


# Loop through each image in the folder
for image_name in os.listdir(image_folder_path):
    image_path = os.path.join(image_folder_path, image_name)

    # Skip directories, only process files
    if os.path.isdir(image_path):
        continue

    # Check if the file is a valid image
    if not image_name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
        print(f"Skipping non-image file: {image_name}")
        continue

    try:
        # Preprocess the image
        result = preprocess_image(image_path)

        # Ensure that the function returns exactly 3 processed images
        if len(result) != 3:
            print(f"Skipping {image_name}: Invalid number of return values")
            continue  # Skip this image if it doesn't return 3 values

        resized, blurred, edges = result

        # Save processed images back to Google Drive
        cv2.imwrite(os.path.join(output_dir, f"{image_name}_resized.jpg"), resized)
        cv2.imwrite(os.path.join(output_dir, f"{image_name}_blurred.jpg"), blurred)
        cv2.imwrite(os.path.join(output_dir, f"{image_name}_edges.jpg"), edges)

        # Increment processed count
        processed_count += 1

        # Print progress every 100 images
        if processed_count % 100 == 0 or processed_count == len(os.listdir(image_folder_path)):
            print(f"Processed {processed_count}/{len(os.listdir(image_folder_path))} images.")

    except ValueError as e:
        print(f"Error processing {image_name}: {e}")
        continue  # Skip the current image if there's an error

# End timer
end_time = time.time()

# Calculate and print the time taken
total_time = end_time - start_time
print(f"Processing completed. {processed_count}/{len(os.listdir(image_folder_path))} images processed successfully.")
print(f"Total time taken: {total_time:.2f} seconds")
```
